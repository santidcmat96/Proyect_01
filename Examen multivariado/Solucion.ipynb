```python
#defino una funcion que me dice cual la distribucion de mi variable
def freq_discrete(df,columnas):
    for c in columnas:
        x=df[c].value_counts().to_frame()
        nombre=x.columns.tolist()[0]
        x.rename(columns={nombre:'Freq'},inplace=True)
        x['Pct']=x['Freq']/x.Freq.sum()
        x['Freq Acum'],x['Pct Acum']=x.Freq.cumsum(),x.Pct.cumsum()
        x.sort_values(['Freq'],ascending=[0],inplace=True)
        print ('Tabla de distribuciÃ³n de Frecuencias para la variable: '+nombre)
        print ('/n')  
        print (x)
        print ('')
def missingss(DF,columnas):
    columnas_missings=[]
    for c in columnas:
        Pct_missings=DF[DF[c].isnull()].shape[0]/DF.shape[0]
        print ('la variable %s tiene un porcentaje de missings de %d'%(c,Pct_missings))
        if Pct_missings>.30:
            columnas_missings.append(c)
    return columnas_missings
def dummys(df,columnas):
    for c in columnas:
        X=df[c]
        for i in list(set(list(df[c]))):
            l=[]
            for j in list(df[c]):
                if j==i:
                    t=1
                elif j!=i:
                    t=0
                l.append(t)
            df[c +' '+ str(i)]=l[:]
    return df
```

```python
train.drop(borrar,axis=1,inplace=True)
test.drop(borrar,axis=1,inplace=True)
```

```python
train=train.loc[train.target==train.target]
```

```python
for f in train.columns:
    train[[f]]=train[[f]].fillna(train[f].mean())
for f in test.columns:
    test[[f]]=test[[f]].fillna(test[f].mean())
```

```python
combinaciones=pd.DataFrame()
combinaciones1=pd.DataFrame()
for k in range(0,len(ventas)):
    for i in range(0,len(ventas)):
        T=train[ventas[5-i]]-train[ventas[k]]
        T1=test[ventas[5-i]]-test[ventas[k]]
        combinaciones=pd.concat([combinaciones,T],axis=1)
        combinaciones1=pd.concat([combinaciones1,T1],axis=1)
        combinaciones.rename(columns={0:'v_combinacion'+str(k)+str(i)},inplace=True)
        combinaciones1.rename(columns={0:'v_combinacion'+str(k)+str(i)},inplace=True)
combinaciones.drop(ventas,axis=1,inplace=True)
combinaciones1.drop(ventas,axis=1,inplace=True)
```

```python
train[list(combinaciones.columns)]=combinaciones
test[list(combinaciones1.columns)]=combinaciones1
```
